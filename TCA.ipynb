```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_csv('Customer Churn.csv')
df.head()



```

```python
df.info()
```

#replacing blanks with 0 as tenure is 0 and no total charges recoded

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
df.info()
```

```python
df.isnull().sum().sum()
```

```python
df["customerID"].duplicated().sum()
```

```python
def conv(value):
  if value == 1:
    return "yes"
  else:
    return "no"

df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
```

#convert 0 and 1 values of senior citizen to yes or no to make it easier to understand


```python
df.head(10)
```

```python
plt.figure(figsize=(3,4))
ax = sns.countplot(x = 'Churn', data = df , hue= "Churn")

ax.bar_label(ax.containers[0])
plt.title("Count of customer by churns")
plt.show()
```

```python
plt.figure(figsize=[3,4])
gb = df.groupby('Churn').agg({'Churn':"count"})

plt.pie(gb['Churn'] , labels = gb.index, autopct = "%1.2f%%")
plt.title("Percentage of churned customers", fontsize = 12)
plt.show()
```

# from thus given pie char we can conclude thta 26.54 customers are churned out

```python
plt.figure(figsize=(3,4))
sns.countplot(x= "gender", data=df , hue= "Churn")
plt.title("Gender of the churned ")
plt.show()
```

```python
plt.figure(figsize=(3,4))
ax = sns.countplot(x= "SeniorCitizen", data = df )
ax.bar_label(ax.containers[0])
plt.title("Count of no. by Senior Citizen")
plt.show()
```

```python


ct = pd.crosstab(df['SeniorCitizen'], df['Churn'], normalize='index') * 100

ax = ct.plot(kind='bar', stacked=True, figsize=(3,4))

for i in range(len(ct)):
    bottom = 0
    for col in ct.columns:
        value = ct.loc[ct.index[i], col]
        ax.text(i, bottom + value/2, f'{value:.1f}%',
                ha='center', va='center')
        bottom += value

plt.title("Churn % among Senior Citizens")
plt.xlabel("Senior Citizen (0 = No, 1 = Yes)")
plt.ylabel("Percentage")
plt.legend(title="Churn")
plt.show()
```

```python
cols = ['PhoneService','MultipleLines','InternetService',
        'OnlineSecurity','OnlineBackup','DeviceProtection',
        'TechSupport','StreamingTV','StreamingMovies']

fig, axes = plt.subplots(3,4 , figsize=(12,10))

for i, col in enumerate(cols):
    ax = axes[i//3, i%3]
    sns.countplot(x=col, data=df, ax=ax , hue = df["Churn"])
    ax.set_title(col)
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

#Online Backup and Device Protection are key drivers of customer loyalty, showing significantly lower churn rates than average. In contrast, basic communication and entertainment services, such as Phone Service and Streaming TV, do not notably improve retention and are often linked to more price-sensitive plans. Ultimately, customers subscribing to security-focused bundles are the most stable, while those with only entertainment or phone services are more likely to leave
